```python
plt.style.use('default')

fig = plt.figure()
x = np.linspace(1,49)
plt.plot(x, 3*x, color='blue', linewidth=1)
plt.xlim(0,50)
plt.ylim(0,160)
plt.title("Draw a line.", fontsize = 10)
plt.xlabel("x - axis")
plt.ylabel("y - axis")
```

```python
from matplotlib.ticker import FormatStrFormatter
fig2 = plt.figure()
fig2, ax = plt.subplots()

plt.title("Sample graph!", fontsize = 12)
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.xlim(1.0,3.0)
plt.ylim(1.0,4.0)
x1 = np.linspace(1,2,2)
x2 = np.linspace(2,3,2)
y1=2*x1
y2=-3*x2+10
plt.plot(x1, y1, color='blue', linewidth=1)
plt.plot(x2, y2, color='blue', linewidth=1)

ax.set_xticks([1, 2, 3])
ax.set_yticks([1, 2, 3, 4]);

ax.xaxis.set_major_formatter(FormatStrFormatter('%.1f')) #sets the ticks format to one decimal place.
ax.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
```
