```python
df = pd.read_csv('olympics.csv', header=1)
df.head()
```

```python
Names =  {'Unnamed: 0': 'Country',
               '? Summer': 'Summer Olympics',
               '01 !': 'Gold',
               '02 !': 'Silver',
               '03 !': 'Bronze',
               '? Winter': 'Winter Olympics',
               '01 !.1': 'Gold.1',
               '02 !.1': 'Silver.1',
               '03 !.1': 'Bronze.1',
               '? Games': '# Games',
               '01 !.2': 'Gold.2',
               '02 !.2': 'Silver.2',
               '03 !.2': 'Bronze.2'}

df.rename(columns = Names, inplace=True)
df.head()
```

```python
df['Country'].is_unique
```

```python
df.isnull().sum()
```

```python
df.sort_values(by='Combined total', ascending=False, inplace=True)
df.head()
```

```python
def medal_count(item):
    if item >= 100:
        return 'Lots of medals'
    if item < 100:
        return 'Few medals'
```

```python
Medals =  df['Combined total'].map(medal_count)
Medals
```
