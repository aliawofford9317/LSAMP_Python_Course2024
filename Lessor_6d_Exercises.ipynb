# **Exercise 2**
Using a pie chart, plot the mass of the planets. You will need to search the Matplotlib documentation on how to create a pie chart.
Add a title "Mass of planets".

```python
#Pie plot of all planets
fig, ax = plt.subplots()
ax.pie(Data['MassKg'],labels=Data['Body'],autopct='%1.1f%%', colors=['purple', 'blue', 'green', 'red'])
ax.axis('equal')
plt.title('Mass of Planets')
plt.ylabel('Mass (kg)')
plt.xlabel('Planets');
```

```python
#Pie plot without Jupiter and Saturn
Mass = Data[0:4]
fig, ax = plt.subplots()
ax.pie(Mass['MassKg'],labels=Mass['Body'], autopct='%1.1f%%')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Mass of Planets')
plt.ylabel('Mass (kg)')
plt.xlabel('Planets');
```

# **Exercise 3**
* Read the births.csv dataset using pandas.
* Plot all the births by month for the year 1992. Annotate the lowest births number and the highest births number on your plot.
* Create a stacked subplot. The top plot should contain the yearly births for the year 2000. The lower subplot should contain all the births for the year 2003.
* Plot both 2000 and 2003 births but this time on a single plot. Add a legend, title, and axis label to this plot.

```python
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
from matplotlib.patches import Circle
from matplotlib.patheffects import withStroke

Births = pd.read_csv('births.csv')
Year = Births.loc[Births['year'] == 1992]
```

```python
plt.scatter(Year['month'],Year['births'])
plt.title('Births per month')
plt.ylabel('Births number')
plt.xlabel('Months');

plt.xlim(0,13)
plt.xticks(range(13))

plt.text(2, 155000, "Lowest birth \nnumber",ha='center', **{'fontsize': 9, 'color': 'black'})
plt.text(7, 181500, "Highest birth \nnumber",ha='center', **{'fontsize': 9, 'color': 'black'})
```

```python
B2000 = Births.loc[Births['year'] == 2000]
B2003 = Births.loc[Births['year'] == 2003]
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], ylim=(150000, 189000))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(150000, 189000))

ax1.set_xlim(0.5, 12.5)
ax2.set_xlim(0.5, 12.5)
plt.xticks(range(13))

ax1.set_title('Yearly births for 2000')
ax2.set_title('Yearly births for 2003')

ax1.plot(B2000['month'],B2000['births'], '.', color='#B629DA')
ax2.plot(B2003['month'],B2003['births'], '.', color=(0.5, 0.7, 1))

```

```python
plt.scatter(B2000['month'],B2000['births'], color='#B629DA', alpha=0.7, label='Year 2000')
plt.scatter(B2003['month'],B2003['births'], color='#0005B4', alpha=0.3, label='Year 2003')
plt.legend(loc='upper left')

plt.title('Births per month')
plt.ylabel('Births number')
plt.xlabel('Months');

plt.xlim(0,13)
plt.xticks(range(13));
```
